```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

## Merging two dataframes by title column

```python
movies=movies.merge(credits,on='title')
movies.head(2)
```

### columns to be retained: generes, id, keywords, overview, title, cast, crew

```python
movies=movies[['id','title','genres','keywords','overview','cast','crew']]
movies.sample(2)
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies.iloc[0]['genres']
```

```python
# converts the 'genres' columns into list format

def convert_genres(genres):
    my_list=[]
    import ast
    genres=ast.literal_eval(genres) # convert string of list into list
    for i in genres:
        my_list.append(i["name"])
    return my_list

movies['genres']=movies['genres'].apply(lambda x:convert_genres(x))
movies['genres']
```

```python
### In the same way we can transform other columns as well
```

```python
movies.iloc[0]['keywords']
```

```python
# converts the 'keywords' columns into list format

def convert_keywords(keywords):
    my_list=[]
    import ast
    keywords=ast.literal_eval(keywords) # convert string of list into list
    for i in keywords:
        my_list.append(i["name"])
    return my_list

movies['keywords']=movies['keywords'].apply(lambda x:convert_keywords(x))
movies['keywords']
```

```python
# converts the 'cast' columns into list format of 3 elements(top actors/actmy_lists)

def convert_cast(cast):
    import ast
    top_cast=ast.literal_eval(cast)
    my_list=[]
    counter=0
    for i in (top_cast):
        if counter>=3:
            break
        my_list.append(i['name'])
        counter+=1
    return my_list

movies['cast']=movies['cast'].apply(convert_cast)
movies['cast']
```

```python
# converts the 'crew' columns into list format of 3 elements(top actors/actmy_lists)

def convert_crew(crew):
    import ast
    top_crew=ast.literal_eval(crew)
    my_list=[]
    for i in (top_crew):
        if i['job']=='Director':
            my_list.append(i['name'])
    return my_list

movies['crew']=movies['crew'].apply(lambda x:convert_crew(x))
movies['crew']
```
