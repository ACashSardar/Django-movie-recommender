```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('movie_data.csv')
movies.drop(columns=['Unnamed: 0'],inplace=True)
```

### Making the overview column from string to list of words

```python
movies['overview']=movies['overview'].apply(lambda x: x.split(" "))
movies
```

### removing the spaces between the names



```python
def remove_space(name):
    return "".join(str(v) for v in name.split(" "))
import ast

movies['genres']=movies['genres'].apply(lambda x: ast.literal_eval(remove_space(x)))
movies['keywords']=movies['keywords'].apply(lambda x: ast.literal_eval(remove_space(x)))
movies['overview']=movies['overview'].apply(lambda x:  ast.literal_eval(remove_space(str(x))))
movies['cast']=movies['cast'].apply(lambda x: ast.literal_eval(remove_space(x)))
movies['crew']=movies['crew'].apply(lambda x: ast.literal_eval(remove_space(x)))
movies.head(5)
```

### creating a 'tag' column that combines genres, keywords, cast, crew 

```python

movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies
```

```python
new_df=movies[['id','title','tag']]

# converting tag column into string
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
new_df.to_csv('clean_data.csv')
```

```python
new_df['tag'].iloc[0]
```

### convert the entire paragraph into lowercase

```python
new_df['tag']=new_df['tag'].apply(lambda x: x.lower())
new_df['tag'].iloc[0]
```

```python
new_df
```

### stemming the text to replace similar type of words with root word

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tag']=new_df['tag'].apply(lambda x: stem(x))
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
vectors
```

### most frequent 5000 words

```python
# cv.get_feature_names()
```

```python
### Finding the cosine similarities
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity
```

```python
def recommend(movie_name):
    indx=movies[movies['title']==movie_name].index[0]
    cos_sim=similarity[indx]
    enum_list=list(enumerate(cos_sim))
    score_list=sorted(enum_list,reverse=True,key=lambda x:x[1])
    score_list=score_list[1:11]

    for i in score_list:
        print(movies['title'].iloc[i[0]])
recommend('Batman v Superman: Dawn of Justice')
```
